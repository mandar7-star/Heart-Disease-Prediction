```python
X = df.drop('target', axis=1)
y = df['target']
```

#### MinMax Scaler (Normalization)

```python
# Create instance of MInMax Scaler (Normalization)
minmax_scaler = MinMaxScaler()
X_scaled = minmax_scaler.fit_transform(X)
X_scaled
X_minmax_df = pd.DataFrame(X_scaled, columns=X.columns)
X_minmax_df
```

#### Standrad Scaler (Standradization)

```python
std_scaler = StandardScaler()
X_scaled = std_scaler.fit_transform(X)
X_scaled
X_std_df = pd.DataFrame(X_scaled, columns=X.columns)
X_std_df
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_minmax_df, y, 
                                            test_size=0.20, random_state=42)
```

```python
X_train
```

```python
X_test
```

### Train Model

```python
knn_class = KNeighborsClassifier(n_neighbors=11, p=2)
knn_class.fit(X_train, y_train)
```

```python
y_pred = knn_class.predict(X_test)
y_pred
```

```python
y_test[:5]
```

```python
y_pred[:5]
```

#### Evaluation

```python
#Confusion Matrix
con_matrix = confusion_matrix(y_test, y_pred)
con_matrix
```

```python
(26+24)/61
```

```python
#Accuracy
accuracy= accuracy_score(y_test, y_pred)
accuracy
```

```python
# Classification Report
class_report = classification_report(y_test, y_pred)
print(class_report)
```

#### Correlation Heatmap

#### Heart Disease Distribution

```python
plt.figure(figsize=(5,4))

sns.countplot(x="target", data=df)

plt.title("Heart Disease Distribution")
plt.xlabel("Target (0 = No, 1 = Yes)")
plt.ylabel("Count")

plt.show()
```

#### Gender vs Heart Disease

```python
plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='target', data=df)
plt.title("Gender vs Heart Disease")
plt.xlabel("0 = Female, 1 = Male")
plt.ylabel("Count")
plt.legend(title="Target", labels=["No Disease", "Disease"])
plt.show()
```

#### Cholesterol vs Heart Disease

```python
plt.figure(figsize=(6,4))

# No disease
plt.scatter(df[df["target"]==0]["chol"],
            df[df["target"]==0]["target"],
            label="No Disease")

# Disease
plt.bar(df[df["target"]==1]["chol"],
            df[df["target"]==1]["target"],
            label="Disease")

plt.xlabel("Cholesterol")
plt.ylabel("Heart Disease (0/1)")
plt.title("Cholesterol vs Heart Disease")

plt.legend()
plt.show()
```
